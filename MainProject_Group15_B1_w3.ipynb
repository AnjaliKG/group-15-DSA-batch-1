# Exploratory Data Analysis

```python
# Distribution graphs (histogram/bar graph) of column data (Univariate)
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]] # For displaying purposes, pick columns that have between 1 and 50 unique values
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) / nGraphPerRow
    plt.figure(num = None, figsize = (2 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} ')
    plt.tight_layout(pad = 0.8, w_pad = 5.0, h_pad = .25)
    plt.show()
    
```

```python
rf_reg = RandomForestRegressor(n_estimators=150, random_state=42)
rf_mod = rf_reg.fit(x_train, y_train)
```

```python
rf_y_pred = rf_mod.predict(x_test)
```

```python
print("Random Forest MSE:", mean_squared_error(y_test, rf_y_pred))
print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_test, rf_y_pred)))
print("Random Forest MAE:", mean_absolute_error(y_test, rf_y_pred))
print("Random Forest R2 Score:", r2_score(y_test, rf_y_pred))
```

```python
dt2 = DecisionTreeRegressor(random_state=1)

dt2.fit(x_train,y_train)
y_pred=dt2.predict(x_test)
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred)))
print("MAE",mean_absolute_error(y_test,y_pred))
print('R-squared:',r2_score(y_test,y_pred)) 
```

```python
adaboost = AdaBoostRegressor(
    RandomForestRegressor(max_depth=5,criterion='mse'),
    n_estimators=150
)
adaboost.fit(x_train, y_train)
```

```python
y_pred = adaboost.predict(x_test)
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred)))
print("MAE",mean_absolute_error(y_test,y_pred))
print('R-squared:',r2_score(y_test,y_pred))
```
